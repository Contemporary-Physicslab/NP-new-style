Meetreeks aan siliconenolie en petroleumolie (parafinneolie) t.b.v. Millikan experiment.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# Siliconenolie

V_so = np.array([0, 1.9, 1.7, 1.55, 1.40, 1.10, 1.0, .75, .55, .4, .25]) #mL
m_so = np.array([3.48, 5.56, 5.36, 5.20, 5.00, 4.73, 4.54, 4.36, 4.12, 3.94, 3.80 ])-3.48 #g

print(m_so[V_so>0]/V_so[V_so>0])
```

```python
# Petroleumolie

V_po = np.array([4.4, 4.5, 5.3, 6.8, 7.2, 9.3, 9.4, 11.6, 13.9, 14.6, 16.9, 18.9]) #mL
m_po = np.array([3.40, 3.56, 3.95, 5.30, 5.57, 7.15, 7.36, 9.06, 10.89, 11.57, 13.38, 14.9 ]) #g
```

```python
def rho(p, V):
    return p*V

popt_so, pcov_so = curve_fit(rho, V_so, m_so)
popt_po, pcov_po = curve_fit(rho, V_po, m_po)

x_test_so = np.linspace(0, 1.1*max(V_so), 100)
x_test_po = np.linspace(0, 1.1*max(V_po), 100)
m_test_so = rho(x_test_so, *popt_so)
m_test_po = rho(x_test_po, *popt_po)


plt.figure()
plt.xlabel('Volume (mL)')
plt.ylabel('Massa (g)')

plt.plot(V_so, m_so, 'r.', label='Siliconenolie')
plt.plot(x_test_so, m_test_so, 'r--', label='Fit: $\\rho = %.2f +/- %.2f$ g/mL ' % (popt_so, np.sqrt(pcov_so[0,0])))

# plt.plot(V_po, m_po, 'b.', label='Petroleumolie')
# plt.plot(x_test_po, m_test_po, 'b--', label='Fit: $\\rho = %.3f +/- %.3f$ g/mL' % (popt_po, np.sqrt(pcov_po[0,0])))

plt.legend()
plt.show()
```
